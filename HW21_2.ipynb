##21.4 Simpson's Rule

Simpsons rule approximates the area under f(x) over two subintervals [xi-1, xi] and [xi,xi+1] by fitting a polynomial though points (xi−1,f(xi−1)), (xi,f(xi)), and (xi+1,f(xi+1)) and integregrading the poynomial

```python
import numpy as np
import matplotlib.pyplot as plt

#definding bounds
a = 0
b = np.pi
n = 11
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)

#using cos function
f = np.cos(x)

#simpson rule integration
I_simp = (h / 3) * (f[0] + 2 * sum(f[2:n-1:2]) + 4 * sum(f[1:n-1:2]) + f[n-1])
err_simp = 0 - I_simp
print("Simpson's Approximation:", I_simp)
print("Error:", err_simp)

#vizualization
x_vals = np.linspace(a, b, 500)
f_vals = np.cos(x_vals)
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='f(x) = cos(x)', color='blue')
plt.plot(x, f, 'ko', label='Sample Points')
plt.grid(True)
plt.legend()
plt.show()
```

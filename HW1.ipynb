##Linear Spaces

Key Terms:

Linear combination: A vector constructed formed by multiplying each vector in a subset and adding the results.

Linear subspace: The result of a linear combination. A linear subspace V is a subset U⊆V which is closed under vector addition and scalar multiplication. Suppose u1, u2 ∈ U and α ∈ R, then


u1 + u2 ∈ U, and α u1 ∈ U.

Linear Independence: A set of vectors that cannot be written as a linear combination of eachother.

Linear Dependence: A set of vectors that can be written as a linear combination of eachother


```python
#linear spaces example
import numpy as np
import sympy as sp


#define vectors v1 and v2
v1 = np.array([1, 2])
v2 = np.array([3, 4])

#linear combination of v1 and v2
a = 2
b = -1
linear_combination = a * v1 + b * v2

#printing results
print("Vector v1:", v1)
print("Vector v2:", v2)
print(f"Linear combination (2*v1 - 1*v2): {linear_combination}")

#checking linear indpednence

#combinding vectors in matrix
matrix = sp.Matrix(np.column_stack((v1, v2)))
#solving rref with sympy function
rref, _ = matrix.rref()
#computing rank of matrix
rank = matrix.rank()

#printing information
print(f"Original Matrix {matrix}")
print(f"RREF of the matrix: {rref}")
print(f"Rank of the matrix: {rank}")

#if rank = num vectors then its linearly indpendent
if rank == matrix.shape[1]:
    print("The vectors are linearly independent. rank = num vectors.")
else:
    print("The vectors are linearly dependent. rank != num vectors")
```

```python
import numpy as np

#define vectors v1 and v2
v1 = np.array([1, 2])
v2 = np.array([3, 4])

print(f"Vector v: {v1}")
print(f"Vector u: {v2}")

#compute dot procut
dot_product = np.dot(v1, v1)

print(f"Dot product: {dot_product}")

#if dot = 0 vectors are orthogonal
if dot_product == 0:
    print("The vectors are orthogonal. dot = 0")
else:
    print("The vectors are not orthogonal. dot != 0")

# Compute the orthogonal projection of v onto u
projection = (np.dot(v1, v2) / np.dot(v1, v2)) * v2

print(f"Orthogonal Projection of v onto u: {projection}")
```

##Eigenvalues and EigenVectors

Let A ∈ R be a square matrix. Then λ ∈ R is an eigenvalue of A if there exists a nonzero vector x such that

Ax = λx

and x is called an eigenvector.

Finding eigenvalues involes solving the characteristic equation

det(A−λI) = 0

Where:

I is the identity matrix of size(A)

λ represents the eigenvalues

After the eigenvalues are found, eigenvectors are then computed with the following equation:

(A−λI)x=0

Where x represents the eigenvectors


```python
import numpy as np
import matplotlib.pyplot as plt

#defining square matrix
A = np.array([[4, -2], [1,  1]])

#computing eigenvalues/eigenvectors with library function
eigenvalues, eigenvectors = np.linalg.eig(A)

#prints results
print(f"Matrix A: {A}")
print(f"\nEigenvalues: {eigenvalues}")
print(f"\nEigenvectors: {eigenvectors}")

#visuzliaztion original basis is in gray and eigenvectors and blue and red

#origin matrix to plot vectors
origin = np.array([[0, 0], [0, 0]])
#plots original basis with respect to origin
plt.quiver(*origin, [1, 0], [0, 1], color=['gray', 'gray'], angles='xy', scale_units='xy', scale=1, alpha=0.5)
#plots eigenvectors with respect to origin
plt.quiver(*origin, eigenvectors[0, :], eigenvectors[1, :], color=['r', 'b'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
